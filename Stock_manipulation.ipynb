```python
import yfinance as yf

ticker_symbol = "TSLA"
tesla = yf.Ticker(ticker_symbol)
tesla_data = tesla.history(period="1y")  
print(tesla_data.head())
```

```python
print(tesla_data.tail())
```

```python
ticker_symbol2 = "GME"
gamestop = yf.Ticker(ticker_symbol2)
gamestop_data = gamestop.history(period="1y")
print(gamestop_data.head())
```

```python
print(gamestop_data.tail())
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], label='Close Price', color='b')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title(title)  # Add a title to the graph
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the make_graph function with Tesla's stock data and a title
make_graph(tesla_data, title="Tesla Stock Price Over the Last Year")
```

```python
make_graph(gamestop_data, title="Tesla Stock Price Over the Last Year")
```
